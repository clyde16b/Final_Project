```python
def Ville_region (VILLE):
    if (VILLE=="Alger" or VILLE=="bedjaia" or VILLE=="Chlef"):
        return "Centre"
    elif (VILLE=="setif" or VILLE=="Constantine"):
        return "Est"
    elif (VILLE=="Oran" or VILLE=="Tlemcen"):
        return "Ouest"

def Tranche_age (Age):
    if (Age>=12 and Age<=24):
        return "12-24 ans"
    elif (Age>=25 and Age<=34):
        return "25-34 ans"
    elif (Age>=35 and Age<=44):
        return "35-44 ans"
    elif (Age>=45 and Age<=54):
        return "45-54 ans"
    elif (Age>=55):
        return "55 ans et plus"
    
def CSP (CSP):
    if (CSP=="Haut fonctionnaire, Cadre supérieur de l'état, …"	or CSP=="Profession libérale, Prof intellectuelle, …"	or CSP=="Cadre"	or CSP=="Cadre moyen / Cadre de maitrise"	or CSP=="Employé/Fonctionnaire"	or CSP=="Petit patron, commerçant, autre indépendant"	or CSP=="Ouvrier, manoeuvre, artisan"	or CSP=="Chômeur"):
        return "Actif"
    elif (CSP=="Femme au foyer"):
        return "Femme au foyer"
    elif (CSP=="Étudiant /écolier"):
        return "Étudiant /écolier"
    elif (CSP=="Retraité"):
        return "Retraité"
    elif (CSP=="Pensionné"):
        return "Pensionné"
    elif (CSP=="Autre inactif"):
        return "Autre inactif"
     
df['Region']=df['VILLE'].map(Ville_region)   
df['Tranche_Age']=df['Age'].map(Tranche_age)
df['CSP_grouped']=df['CSP'].map(CSP)
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

# Préparation des X et des y

```python
X=df.drop(["Tranche_Age","Region","CSE","Sexe","VILLE","NUMQSTR","Age",'CSP_grouped','CSP','Niveau','Statut_Familial','Situation_Matrimoniale'],axis=1)
y=df["CSE"]
```

# Spliter la Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test=train_test_split(X,y, test_size=.25, random_state=20)
```

# Méthode des k plus proche voisin

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score,precision_score,recall_score
```

```python
from math import sqrt
sqrt(len(X_test))
```

```python
knclf=KNeighborsClassifier(n_neighbors=16)
knclf.fit(X_train,y_train)
y_pred_KNN=knclf.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred_KNN)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt=".1f")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title("Matircie de confusion modèle KNN")
plt.show()
```

```python
scr_KNN=knclf.score(X_test,y_test)*100
accscr_KNN=accuracy_score(y_test,y_pred_KNN)*100
f1scr_KNN=f1_score(y_test,y_pred_KNN,average="macro")*100
pescr_KNN=precision_score(y_test,y_pred_KNN,average="macro")*100
rescr_KNN=recall_score(y_test,y_pred_KNN,average="macro")*100

print("Les metrics: \n"
     "\t score=",scr_KNN,"%\n",
     "\t accuracy_score=",accscr_KNN,"%\n",
     "\t f1_score=",f1scr_KNN,"%\n",
     "\t precision_score=",pescr_KNN,"%\n",
     "\t recall_score=",rescr_KNN,"%\n",)
```

# Méthode Arbre de décision

```python
from sklearn.tree import DecisionTreeClassifier

dectre=DecisionTreeClassifier(criterion = 'gini', random_state = 50)
dectre.fit(X_train, y_train)
y_pred_DT = dectre.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred_DT)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt=".1f")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title("Matircie de confusion modèle Arbre de décision")
plt.show()
```

```python
scr_DT=dectre.score(X_test, y_test)*100
acc_DT=accuracy_score(y_test,y_pred_DT)*100
f1_DT=f1_score(y_test,y_pred_DT,average="macro")*100
rec_DT=recall_score(y_test,y_pred_DT,average="macro")*100
pre_DT=precision_score(y_test,y_pred_DT,average="macro")*100

print("Les metrics: \n"
     "\t score=",scr_DT,"%\n",
     "\t accuracy_score=",acc_DT,"%\n",
     "\t f1_score=",f1_DT,"%\n",
     "\t precision_score=",rec_DT,"%\n",
     "\t recall_score=",pre_DT,"%\n",)
```
